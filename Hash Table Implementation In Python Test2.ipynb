 Hash tables are a type of data structure in which the address or the index value of the data element is generated from a hash function. That makes accessing the data faster as the index value behaves as a key for the data value. In other words Hash table stores key-value pairs but the key is generated through a hashing function.

In this case, we shall use the python in built dictionary data type to show case tables since it stores in key-value pairs

The keys of the dictionary are hashable i.e. the are generated by hashing function which generates unique result for each unique value supplied to the hash function.

The keys in the dictionary should be:
    1.Hashable(They should be generated by the hashing function producing a unique value for every input.)
    2.The order of the data elements in a dictionary is not fixed

Perfect hashing or perfect hash function is the one which assigns a unique slot for every key value.

```python
# let me create hash table of size 101
hash_table = [None] * 10
print (hash_table) 
```

Modulo operator (%) is used in the hashing function. 
The % (modulo) operator yields the remainder from the division of the first argument by the second.

```python
#we first create the hash function 
def hash_function(key):
    return key % len(hash_table)
 
print (hash_function(10))
print (hash_function(20))
print (hash_function(25))    
```

To insert some data in the table, 
we first use the hashing function to generate a slot/index and store the given value into that slot.

```python
def insert(hash_table, key, value):
    hash_key = hash_function(key)
    hash_table[hash_key] = value 
 
insert(hash_table, 10, 'Uganda')  
insert(hash_table, 25, 'Kenya')
print (hash_table)
```

```python
#collisions
#A collision occurs when two items/values get the same slot/index.
insert(hash_table, 20, 'India')#Replaces Uganda with India
print (hash_table)
```

# Chaining

```python
#Chaining helps us store the excess value in the open slot within a hash table.

hash_table = [[] for _ in range(10)]
print (hash_table)
```

```python
def hashing_func(key):
    return key % len(hash_table)

print (hashing_func(10)) # Output: 0
print (hashing_func(20)) # Output: 0
print (hashing_func(25)) # Output: 5
```

```python
# We change the insert function. We use append() function to insert key-value pairs in the hash table.

def insert(hash_table, key, value):
    hash_key = hashing_func(key)
    hash_table[hash_key].append(value)

insert(hash_table, 10, 'Nepal')
print (hash_table)
# Output: 
# [['Nepal'], [], [], [], [], [], [], [], [], []]
```

```python
insert(hash_table, 20, 'India')
print (hash_table)
```
